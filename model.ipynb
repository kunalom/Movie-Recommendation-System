```python
credits=pd.read_csv('/kaggle/input/tmdb-movie-metadata/tmdb_5000_credits.csv')
movies=pd.read_csv('/kaggle/input/tmdb-movie-metadata/tmdb_5000_movies.csv')
```

```python
movies=movies[['id', 'title', 'keywords', 'genres', 'overview']]
```

```python
credits.head()
```

```python
credits.rename(columns={'movie_id':'id'}, inplace=True)
credits.head(1)
```

```python
credits.shape
```

```python
movies=movies.merge(credits, on=['id', 'title'])
movies.head(1)
```

```python
movies.shape
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
movies.shape
```

```python
movies.iloc[0].genres
```

```python
def extract(row):
    l=[]
    for i in row:
        l.append(i['name'])
        
    return l
```

```python
movies['genres'] = movies['genres'].apply(lambda x: extract(eval(x)))
movies.head(3)
```

```python
movies.iloc[0].keywords
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x: extract(eval(x)))
movies.head(3)
```

```python
def extract(row):
    l=[]
    count=0;
    for i in row:
        l.append(i['name'])
        count+=1
        if(count==3):
          break
        
    return l
```

```python
movies['cast'] = movies['cast'].apply(lambda x: extract(eval(x)))
movies.head(3)
```

```python
def extract(row):
    l=[]
    for i in row:
        if i['job']=='Director':
           l.append(i['name'])
        
    return l
```

```python
movies['crew'] = movies['crew'].apply(lambda x: extract(eval(x)))
movies.head(3)
```

```python
def removespaces(l):
    L=[]
    for i in l:
        L.append(i.replace(" ",""))
    return L
```

```python
movies['cast'] = movies['cast'].apply(removespaces)
movies['crew'] = movies['crew'].apply(removespaces)
movies['genres'] = movies['genres'].apply(removespaces)
movies['keywords'] = movies['keywords'].apply(removespaces)
```

```python
movies.head(3)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head(3)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies = movies.drop(columns=['overview','genres','keywords','cast','crew'])
movies.head(3)
```

```python
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x))
movies.head()
```

```python
movies['tags']=movies['tags'].apply(lambda x:x.lower())
movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector = cv.fit_transform(movies['tags']).toarray()
```

```python
len(cv.get_feature_names_out())
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
def stemming(row):
    l=[]
    for i in row.split():
        l.append(ps.stem(i))
    return " ".join(l)
```

```python
movies['tags']=movies['tags'].apply(stemming)
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
how_much_similar=cosine_similarity(vector)
```

```python
how_much_similar.shape
```

```python
def recommend(movie):
    index=movies[movies['title']==movie].index[0]
    distance=sorted(enumerate(how_much_similar[index]), reverse=True, key=lambda x: x[1])
    for i in distance[1:6]:
        print(movies.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```
